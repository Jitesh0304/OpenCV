# Sketch

```python
import numpy as np
import cv2
```

```python
def sketch_transform(image):
    
    ## convert the image into gray color
    image_grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    ## Blur the image using GaussianBlur
    image_grayscale_blure = cv2.GaussianBlur(image_grayscale, (7,7), 0)
    
    ## Canny is use for edge detection ..... using this method we can draw sketch
    image_canny = cv2.Canny(image_grayscale_blure, 10, 80)
    
    ##
    _, mask = image_canny_inverted = cv2.threshold(image_canny, 30, 255, cv2.THRESH_BINARY_INV)
    
    return mask
```

```python
cam_capture = cv2.VideoCapture(0)
cv2.destroyAllWindows()

while True:
    _, im0 = cam_capture.read()
    showCrosshair = False
    fromCenter = False
    
    r = cv2.selectROI("image", im0 , showCrosshair, fromCenter)
    break
    
    
while True:
    _, image_frame = cam_capture.read()
    
    rect_image = image_frame[int(r[1]) : int(r[1] + r[3]), int(r[0]): int(r[0] + r[2])]
    
    sketcher_rect = rect_image
    
    sketcher_rect = sketch_transform(sketcher_rect)
    
    sketcher_rect_rgb = cv2.cvtColor(sketcher_rect, cv2.COLOR_GRAY2BGR)
    
    image_frame[int(r[1]) : int(r[1] + r[3]), int(r[0]): int(r[0] + r[2])] = sketcher_rect_rgb
    
    cv2.imshow("sketch", image_frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
        
cam_capture.release()
cv2.destroyAllWindows()
```

```python
image = cv2.imread('Modi.jpg',0)
edges = cv2.Canny(image, 100, 200)

plt.figure(figsize= (12,12))
plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title('original_image'), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(edges, cmap= 'gray'), plt.title('Canny_image')
plt.xticks([]), plt.yticks([])
```
