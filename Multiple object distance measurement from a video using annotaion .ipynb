```python
import cv2
import numpy as np
import time
import imutils
from imutils import paths

    ## object distance from camera
KNOWN_DISTANCE = 12

    ## object width in real
KNOWN_WIDTH = 3.3


## some colors
Green = (0,255,0)          
Red = (0,0,255)
Black = (255,255,255)
White = (0,0,0)

        ## find focal length 
def Focal_length_finder(measured_distance, real_width, width_in_ref_image):
    
    focal_length = (width_in_ref_image * measured_distance) / real_width
    
    return focal_length


              ## find distance of the object from the camera
def Distance_finder(Focal_length, real_width_object , object_width_in_frame):

    distace = (real_width_object * Focal_length) / object_width_in_frame

    return distace


        ## draw rectangle on the photo
def make_rectangle(event, x, y, flag, parameters):
    
        ## read the image
    image = cv2.imread('jio_fi1.jpeg')
    
        ## if your image size is too big..convert the size of image into small size
    height, width = image.shape[:2]
    desired_size = (width//3, height//3)
    resized_img = cv2.resize(image, desired_size, interpolation = cv2.INTER_LINEAR)
    
        ## write the logic of mouse_clicking on the image
        ## when ever i click on the image .... the click position will taken into consider
    if event == cv2.EVENT_LBUTTONDOWN:
        
            ## applend the (x,y) coordinate value of the clicked position
            ## append both values simultaniously .... as a set of value
        list1.append((x,y))
        
            ## resize the image as we did before
        height, width = image.shape[:2]

        desired_size = (width//3, height//3)

        resized_img = cv2.resize(image, desired_size, interpolation = cv2.INTER_LINEAR)
        
            ## draw a circle on the clicked position
        cv2.circle(resized_img, center =(x, y), radius = 1, color =(0,0,0), thickness = 1)
        
        
            ## if we clicked 2 times or more on the image
        ## if len(list1) % 2 == 0:
        if len(list1) == 2:
            
                ## draw rectanle between last two clicked positions on the image(clicked_1  &  clicked_2)
            cv2.rectangle(resized_img, pt1 = list1[-1], pt2 =list1[-2], color =(0,255,0), thickness =1)
            
                ## calculate the width and height of the rectangle
            width , height = (list1[-1][0] - list1[-2][0]), (list1[-1][1] - list1[-2][1])
            
                ## calculate the focal length 
                ## (width_of_object in pixels * actual_distance_of_the_object from the camera) / actual width of objext in real
            focalLength = (width * KNOWN_DISTANCE) / KNOWN_WIDTH
            
                ## store the focal_length in a list
            focal_list.append(focalLength)
            
                ## calculate the distance of the object
            inches = Distance_finder(focalLength, KNOWN_WIDTH, width)
            
                ## display the distance of the screen
            cv2.putText(resized_img, "%.2fcm" % (inches * 2.5), (resized_img.shape[1] - 300, resized_img.shape[0] - 100), 
                        cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 0), 3)
            
            
            
            ## display the image after calculating the distance
        cv2.imshow('window1',resized_img)
        
        
    ## read the image
image = cv2.imread('photo1.jpeg')

    ## resize the image
height, width = image.shape[:2]
desired_size = (width//3, height//3)
resized_img = cv2.resize(image, desired_size, interpolation = cv2.INTER_LINEAR)

    ## display the image
cv2.imshow('window1', resized_img)

## derived the list we required
list1 =[]
focal_list = []

    ## call the mouse function
cv2.setMouseCallback('window1', make_rectangle)

cv2.waitKey(0)
cv2.destroyAllWindows()


## The above code is for the referal image
## from the above code we will only consider the focal_length 
## Then we compare that with our new images 

cap = cv2.VideoCapture('video1.mp4')
## Looping the file and process each and every image
while cap.isOpened():
    
    _, frame = cap.read()
    
    Green = (0,255,0)          
    Red = (0,0,255)
    Black = (255,255,255)
    White = (0,0,0)


                  ## distance estimator fuction
    def Distance_finder(Focal_length, real_width_object , object_width_in_frame):

        distace = (real_width_object * Focal_length) / object_width_in_frame

        return distace


    def make_rectangle(event, x, y, flag, parameters):

        if event == cv2.EVENT_LBUTTONDOWN:

            list1.append((x,y))

            cv2.circle(frame, center =(x, y), radius = 1, color =(0,0,0), thickness = 1)


            if len(list1) % 2 == 0:

                cv2.rectangle(frame, pt1 = list1[-1], pt2 =list1[-2], color =(0,255,0), thickness =1)

                width , height = (list1[-1][0] - list1[-2][0]), (list1[-1][1] - list1[-2][1])

                inches = Distance_finder(int(focal_list[0]), KNOWN_WIDTH, width)

                cv2.putText(frame, "%.2fcm" % (inches * 2.5), (list1[-1][0],list1[-1][1]), 
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)
                
            cv2.imshow('window1',frame)
        
        

    _, frame = cap.read()

    cv2.imshow('window1', frame)
    list1 =[]
    cv2.setMouseCallback('window1', make_rectangle)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    if cv2.waitKey(1) == ord("q"):
        break
    
    
## closing 
cap.release()

## closing windows 
cv2.destroyAllWindows()
```
